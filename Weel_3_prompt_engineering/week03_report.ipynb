## Comparative Insights

After completing all tasks using Zero-Shot, Few-Shot, and Chain-of-Thought (CoT) prompting, several clear patterns emerged:

- **Zero-Shot Prompting:**  
  Fastest but often lacks reasoning transparency. Suitable for simple problems.

- **Few-Shot Prompting:**  
  Improves accuracy with structured examples. Works well for pattern-based tasks.

- **Chain-of-Thought (CoT) Prompting:**  
  Significantly improves reasoning clarity by forcing step-by-step logic. Best for multi-step reasoning and explanations.

**Overall:**  
CoT > Few-Shot > Zero-Shot in reasoning quality and reliability.




## Reflection

1. **Accuracy vs Transparency:** CoT offers both correctness and reasoning clarity.  
2. **Influence of Examples:** Few-Shot helps pattern recognition but limits flexibility.  
3. **Reasoning Quality:** CoT improves step articulation and human interpretability.  
4. **Practical Use:**  
   - Zero-Shot → simple Q&A  
   - Few-Shot → pattern-based tasks  
   - CoT → multi-step reasoning  
5. **Key Takeaways:**  
   - CoT > Few-Shot > Zero-Shot  
   - Examples help but don’t replace logical reasoning  
   - Prompt phrasing drastically affects output  

**Conclusion:**  
Chain-of-Thought prompting is the most effective method for reasoning-based tasks, ensuring both **accuracy** and **clarity**.
